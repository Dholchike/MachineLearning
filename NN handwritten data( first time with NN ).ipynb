```
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```
(X_train , y_train ) , ( X_test , y_test) = keras.datasets.mnist.load_data()
```

```

X_train = X_train/255
y_train = y_train/255
```

```
X_train[0]
```

```
# now going to flatten our dataset bcz our dataset is 28*28 ... so we need it in single array ...therefore we'll reshape it
# to 784 

X_train_flattened = X_train.reshape(len(X_train),28*28)
#X_train_flattened.shape
X_test_flattened = X_test.reshape(len(X_test),28*28)
#X_test_flattened.shape
```

```
X_train_flattened[0]
```

```
# now we'll do NN , abhi tak data flatten and file kr rahe the hum
```

```
model = keras.Sequential([keras.layers.Dense(10  , input_shape = (784,), activation = "sigmoid")])  # dense se har neuron har aage wale neuron se conected hota hai])


model.compile( optimizer = "adam"  , loss = "sparse_categorical_crossentropy" , metrics = "accuracy")
model.fit(X_train_flattened , y_train, epochs = 5) # 5= no of iteration kitni baaar chalega NN

```

```
model.evaluate(X_test_flattened,y_test)
```
