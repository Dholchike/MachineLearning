```
import pandas as pd
import numpy as np
import sklearn
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
titanic = pd.read_csv(io.BytesIO(uploaded['titanic.csv']))
```

```
titanic.head()
```

```

# now i noticed that age has some "nan" values , so we neeed to do something 
# so according to the codebasics Naive bayes part 1 video , popularly we take the mean of the column and replace those NaN entries with the mean 
#titanic.Age = titanic.Age.fillna(titanic.Age.mean())
#titanic.head(50)

#doing the above gave 80%ish accuracy , so im trying to drop all rows with NAN values
titanic = titanic.dropna()
```

```
len(titanic)
```

```
y = titanic['Survived']
y.head()
```

```
X = titanic[['Pclass','Sex' , 'Age','Fare']]
X.head(50)
```

```
len(X)
```

```
X.head(50)
```

```
dummies = pd.get_dummies(X.Sex)
dummies.head(4)
```

```
X = pd.concat([X, dummies], axis = 'columns')
X.head()
```

```
#now dropping SEX and Female Column 
X.drop(['Sex',"female"] , axis = "columns" , inplace = True)

X.head(10)
```

```
X = X.astype({"Age":int,"Fare":int})
# convert column "a" to int64 dtype and "b" to complex type
#df = df.astype({"a": int, "b": complex})
X.head(10)
```

```
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.2 , random_state = 1 )
```

```
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```
model.score(X_test , y_test)
```
